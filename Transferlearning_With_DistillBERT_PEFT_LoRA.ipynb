# Import Libraries, Loading data, Pre processing, and splitting data into training, validation and testing data

```python
# Import libraries for Transfer Learning

import pandas as pd
import numpy as np
import torch # PyTorch for deep learning framework such as transformers
from datasets import Dataset, DatasetDict # DatasetDict: Hugging Face datasets for efficient data handling
from transformers import (
    AutoTokenizer,                        # Automatically loads appropriate tokenizer
    AutoModelForSequenceClassification,   # Pre-trained model for classification tasks
    TrainingArguments,                    # Configuration for training parameters
    Trainer,                              
    DataCollatorWithPadding,              # it handles dynamic padding during batching
    EarlyStoppingCallback                 # Stops training when no improvement
)
from peft import LoraConfig, get_peft_model, TaskType  # for Parameter-Efficient Fine-Tuning
from sklearn.model_selection import train_test_split  
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix  # Evaluation metrics
import warnings
warnings.filterwarnings('ignore')  # Suppresses warning messages

print("ENRON SPAM CLASSIFIER WITH DISTILBERT + LoRA") # with LoRA (Low Rank Adaptation)

# LOADING DATA

DATA_PATH = '/kaggle/input/enron-spam-data/enron_spam_data.csv'

try:
    df = pd.read_csv(DATA_PATH) # Load CSV file into pandas DataFrame
    print(f" File loaded successfully")
except FileNotFoundError:
    print(f" File not found: {DATA_PATH}")
    
    raise # Re-raise exception to stop execution

# Preprocessing data
df['Subject'] = df['Subject'].fillna('')      # Replacing missing subjects with empty strings
df['Message'] = df['Message'].fillna('')      # Replacing missing messages with empty strings
df['text'] = (df['Subject'] + ' ' + df['Message']).str.strip()  # Combining subject and message
df['label'] = (df['Spam/Ham'] == 'spam').astype(int)  # Converting to binary: spam=1, ham=0
df = df[df['text'].str.len() > 5].reset_index(drop=True)  # Removing very short emails (to remove noise)

print(f"Dataset: {len(df):,} emails")
print(f" Ham:  {(df['label']==0).sum():,} ({(df['label']==0).sum()/len(df)*100:.1f}%)")
print(f" Spam: {df['label'].sum():,} ({df['label'].sum()/len(df)*100:.1f}%)")

# Creating word_count column 
df['word_count'] = df['text'].apply(lambda x: len(x.split()))

# Removing very long emails (>2000 words) to avoid noise
# Counting emails to be removed
very_long = (df['word_count'] > 2000).sum()
print(f"Removing {very_long:,} emails with >2000 words")

# Apply filter - Removing only emails with more than 2000 words
df = df[df['word_count'] <= 2000].reset_index(drop=True)

# Display statistics after filtering
print(f"\nAfter filtering:")
print(f"Total emails: {len(df):,} (removed {very_long:,} emails)")

# Drop the word_count column 
df = df.drop(columns=['word_count'])

print(f"   Total emails: {len(df):,}")
print(f"   Ham:  {(df['label']==0).sum():,} ({(df['label']==0).sum()/len(df)*100:.1f}%)")
print(f"   Spam: {df['label'].sum():,} ({df['label'].sum()/len(df)*100:.1f}%)")

# Split into 80% train and validation and 20% as test data (unseen data to model)
train_val_df, test_df = train_test_split(
    df[['text', 'label']], 
    test_size=0.20,              # 20% for testing (unseen data: held out completely)
    random_state=42, 
    stratify=df['label']         # Maintains class distribution
)

# Splitting train_val into 90% training and 10% validation (from the 80%)
# 10% of 80% = 8% of total dataset for validation
train_df, val_df = train_test_split(
    train_val_df, 
    test_size=0.125,             # 0.125 * 80% = 10% of original dataset
    random_state=42, 
    stratify=train_val_df['label']
)

print(f"Train:      {len(train_df):,} samples ({len(train_df)/len(df)*100:.1f}% of total)")
print(f"Validation: {len(val_df):,} samples ({len(val_df)/len(df)*100:.1f}% of total)")
print(f"Test:       {len(test_df):,} samples ({len(test_df)/len(df)*100:.1f}% of total)")
```

# LOADING MODEL & TOKENIZER

```python
# Convert to Hugging Face Dataset format (It is more efficient than pandas)
dataset = DatasetDict({
    'train': Dataset.from_pandas(train_df, preserve_index=False),
    'validation': Dataset.from_pandas(val_df, preserve_index=False),
    'test': Dataset.from_pandas(test_df, preserve_index=False)
})


# LOADING MODEL & TOKENIZER

MODEL_NAME = "distilbert-base-uncased" # DistilBERT: lighter, and faster version of BERT
print(f"\n Loading {MODEL_NAME}")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME) # Loads tokenizer for text processing
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=2, # Ham/Spam as a Binary classification problem
    id2label={0: "ham", 1: "spam"}, # Maps numeric IDs to text labels
    label2id={"ham": 0, "spam": 1}  # Maps text labels to numeric IDs
)
print("Model is loaded")
```

# APPLYING LoRA (Low Rank Adaptation)

```python
# LoRA adds small trainable matrices to model layers instead of fine-tuning all parameters

lora_config = LoraConfig(
    task_type=TaskType.SEQ_CLS,  # Sequence Classification task
    r=16,                         # Rank of LoRA matrices (lower means few parameters)
    lora_alpha=32,                # Scaling factor for LoRA weights
    lora_dropout=0.1,             # Dropout rate to prevent overfitting
    target_modules=["q_lin", "v_lin"], # Apply LoRA only to query and value layers
    bias="none"                   # It will not train bias parameters
)

model = get_peft_model(model, lora_config)  # Model with LoRA adapters
model.print_trainable_parameters()  # Reduction in trainable parameters
```

# TOKENIZING DATASET

```python

def tokenize_function(examples):
    """Converting text to numerical tokens that the model can process"""
    return tokenizer(
        examples['text'],
        padding=False,        # Will not do padding here (will do pad dynamically with batches)
        truncation=True,      # Cut off text longer than max_length
        max_length=512        # Maximum sequence length (DistilBERT limit)
    )

print("\n Tokenizing dataset")

# Applying tokenization to entire dataset in batches 
tokenized_dataset = dataset.map(
    tokenize_function,
    batched=True,              # Processing examples at once in batch
    remove_columns=['text'],   # Removing original text (We no longer need now)
    desc="Tokenizing"
)

# Handles dynamic padding: pads sequences to longest in each batch

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
print("Tokenization complete")
```

# FUNCTION FOR EVALUATION METRICS

```python
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1) # Converting probabilities to class predictions
    
    accuracy = accuracy_score(labels, predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, predictions, average='binary'
    )
    
    return {
        'accuracy': accuracy,
        'precision': precision, # Of predicted spam, how many are actually spam
        'recall': recall,        # Of actual spam, how many were caught
        'f1': f1
    }
```

# TRAINING CONFIGURATION 

```python
training_args = TrainingArguments(
    output_dir="./results",  # saving checkpoints
    
    # Training parameters
    num_train_epochs=3,                    # Number of complete passes through training data
    per_device_train_batch_size=16,        # Samples per batch during training
    per_device_eval_batch_size=32,         # Larger batch for evaluation
    
    # Optimization
    learning_rate=2e-4,                    # Step size for weight updates 
    weight_decay=0.01,                     # L2 regularization to prevent overfitting
    warmup_steps=500,                      # Gradually increase learning rate at start
    
    # Evaluation & Checkpointing
    evaluation_strategy="steps",           # Evaluate every N steps
    eval_steps=500,                        # Evaluate every 500 training steps
    save_strategy="steps",                 # Save checkpoint every N steps
    save_steps=500,
    save_total_limit=2,                    # Keeping only 2 most recent checkpoints 
    load_best_model_at_end=True,           # Loading best checkpoint after training
    metric_for_best_model="f1",            # Use F1 score to determine the "best" model
    
    # Performance optimizations
    fp16=True,                             # Using 16-bit floating point (faster, less memory)

    dataloader_num_workers=2,              # Parallel data loading
    gradient_accumulation_steps=2,         # Accumulate gradients over 2 steps
    
    # Logging
    logging_steps=100,                     # Log metrics every 100 steps
    logging_dir='./logs',
    report_to="none",                      
    
    # Reproducibility
    seed=42,                               # Fixed random seed for consistent results
)

```

# CREATING A TRAINER AND TRAINING MODEL

```python

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset['train'], # Only training data (72% of total)
    eval_dataset=tokenized_dataset['validation'], # Validation data (8% of total)
    tokenizer=tokenizer,
    data_collator=data_collator,           # Handles padding
    compute_metrics=compute_metrics,        # Calculate metrics during evaluation
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)]  # Stop if no improvement for 3 evaluations
)


# TRAINING THE MODEL

print(f"Device: {training_args.device}")
print(f"Epochs: {training_args.num_train_epochs}")
print(f"Batch size: {training_args.per_device_train_batch_size}")
print(f"Learning rate: {training_args.learning_rate}")

train_result = trainer.train()  # Executing the training loop

print(" TRAINING COMPLETED")

print(f" Total time: {train_result.metrics['train_runtime']:.1f}s ({train_result.metrics['train_runtime']/60:.1f} min)")
print(f" Throughput: {train_result.metrics['train_samples_per_second']:.1f} samples/sec")
print(f" Final loss: {train_result.metrics['train_loss']:.4f}")

```

# EVALUATION ON VALIDATION SET (During Training)

```python
print("\nVALIDATION SET RESULTS (Used during training for model selection):")
val_results = trainer.evaluate(tokenized_dataset['validation'])
print(f"   Accuracy:  {val_results['eval_accuracy']*100:.2f}%")
print(f"   Precision: {val_results['eval_precision']*100:.2f}%")
print(f"   Recall:    {val_results['eval_recall']*100:.2f}%")
print(f"   F1 Score:  {val_results['eval_f1']*100:.2f}%")

```

```python
# Evaluate on held-out test set
test_results = trainer.evaluate(tokenized_dataset['test'])

print(f"\n   Accuracy:  {test_results['eval_accuracy']*100:.2f}%")
print(f"   Precision: {test_results['eval_precision']*100:.2f}%")
print(f"   Recall:    {test_results['eval_recall']*100:.2f}%")
print(f"   F1 Score:  {test_results['eval_f1']*100:.2f}%")

# CONFUSION MATRIX ON TEST SET (Unseen data)

print("\n predictions on test set")
predictions = trainer.predict(tokenized_dataset['test'])
y_pred = np.argmax(predictions.predictions, axis=1)  # Predicted labels
y_true = predictions.label_ids                        # True labels

cm = confusion_matrix(y_true, y_pred)
tn, fp, fn, tp = cm.ravel()  # True Negative, False Positive, False Negative, True Positive

print("\n CONFUSION MATRIX (Test Set):")
print(f"\n                Predicted")
print(f"              Ham    Spam")
print(f"Actual Ham   {cm[0][0]:5d}  {cm[0][1]:5d}")
print(f"       Spam  {cm[1][0]:5d}  {cm[1][1]:5d}")

print(f"\nDETAILED EVALUATION METRICS (Test Set):")
print(f"   True Positive Rate (Sensitivity/Recall): {tp/(tp+fn)*100:.2f}%")
print(f"   True Negative Rate (Specificity):        {tn/(tn+fp)*100:.2f}%")
print(f"   False Positive Rate:                     {fp/(fp+tn)*100:.2f}%")
print(f"   False Negative Rate:                     {fn/(fn+tp)*100:.2f}%")
print(f"   Positive Predictive Value (Precision):   {tp/(tp+fp)*100:.2f}%")
print(f"   Negative Predictive Value:               {tn/(tn+fn)*100:.2f}%")

# Calculate additional metrics
print(f"\n   Test Set Size: {len(y_true):,} emails")
print(f"   Correctly Classified: {(y_true == y_pred).sum():,} ({(y_true == y_pred).sum()/len(y_true)*100:.2f}%)")
print(f"   Misclassified: {(y_true != y_pred).sum():,} ({(y_true != y_pred).sum()/len(y_true)*100:.2f}%)")

```

## EXTRACT ATTENTION WEIGHTS & HIDDEN STATES
## Function to get model outputs

```python

def get_model_outputs(text, label_name):
    """
    Extracts internal representations from the model
    
    Returns:
        - Attention weights: Which words the model focuses on
        - Hidden states: Internal numerical representations
        - Predictions: Model's confidence scores
    """
    
    # Tokenize input text
    inputs = tokenizer(text, return_tensors="pt", truncation=True, 
                      max_length=128, padding='max_length')  # Shorter max_length for faster processing
    inputs = {k: v.to(device) for k, v in inputs.items()}  # Move to GPU/CPU
    
    # Get model outputs without computing gradients (saves memory)
    with torch.no_grad():
        outputs = model(**inputs, 
                       output_attentions=True,      # Extract attention weights,Shape: [num_layers, num_heads, seq_length, seq_length]
                       output_hidden_states=True)   # Extract hidden representations,Shape: [num_layers, seq_length, hidden_dim]
    
    return {
        'text': text,
        'label': label_name,
        'input_ids': inputs['input_ids'],           # Token IDs
        'attention_mask': inputs['attention_mask'], # Which tokens are real (not padding)
        'attentions': outputs.attentions,           # All attention layers
        'hidden_states': outputs.hidden_states,     # All hidden layer outputs
        'logits': outputs.logits,                   # Raw predictions
        'prediction': torch.softmax(outputs.logits, dim=1).cpu().numpy()[0]  # Probabilities
    }
```

## Process all email pairs

```python

all_outputs = []
for i, pair in enumerate(email_pairs, 1):
    ham_output = get_model_outputs(pair['ham'], 'ham')
    spam_output = get_model_outputs(pair['spam'], 'spam')
    all_outputs.extend([ham_output, spam_output])  # Store both

print(f"Extracted outputs from {len(all_outputs)} emails")
```

## EXTRACT FEATURES FOR VISUALIZATION

```python
num_layers = len(all_outputs[0]['attentions'])  # DistilBERT has 6 layers
last_layer_idx = -1  # Using the last (most sophisticated)layer

attention_vectors = []
hidden_state_vectors = []
labels = []
texts = []
predictions = []

for output in all_outputs:
    # ATTENTION FEATURES 
    # Get attention from last layer: [num_heads, seq_len, seq_len]
    attention = output['attentions'][last_layer_idx][0]
    
    # Average across all attention heads to get single attention pattern
    avg_attention = attention.mean(dim=0)  # [seq_len, seq_len]
    
    # Extract [CLS] token's attention to all other tokens
    # This shows which words are most important for classification
    cls_attention = avg_attention[0, :].cpu().numpy()  # [seq_len]
    attention_vectors.append(cls_attention)
    
    # HIDDEN STATE FEATURES 
    # Get [CLS] token's hidden representation from last layer
    # This is the "summary" representation used for classification
    last_hidden_state = output['hidden_states'][-1][0]  # [seq_len, hidden_dim]
    cls_hidden = last_hidden_state[0, :].cpu().numpy()  # [hidden_dim] (768 dimensions)
    hidden_state_vectors.append(cls_hidden)
    
    # METADATA 
    labels.append(output['label'])
    texts.append(output['text'][:60] + "...")
    predictions.append(output['prediction'])

attention_vectors = np.array(attention_vectors)  # Shape: [20, 128] (20 emails, 128 tokens)
hidden_state_vectors = np.array(hidden_state_vectors)  # Shape: [20, 768] (20 emails, 768 dimensions)


print(f"   Model has {num_layers} attention layers") # confirming attention layers 
print(f"Attention vectors shape: {attention_vectors.shape}") # confirming attention vetors shape 
print(f"Hidden state vectors shape: {hidden_state_vectors.shape}") # confirming hidden state vector shape 

```

## PERFORMING DIMENSIONALITY REDUCTION

```python

# Problem is that , As We Can not visualize 128-dimensional attention and 768-dimensional hidden states
# So, we can reduce to 2D while also preserving the structure 

n_samples = attention_vectors.shape[0]  # 20 emails
n_features_attn = attention_vectors.shape[1]  # 128 dimensions
n_features_hidden = hidden_state_vectors.shape[1]  # 768 dimensions

# PCA (Principal Component Analysis)
# PCA is a Linear dimensionality reduction, it also enables to remove noise
pca_components_attn = min(n_samples - 1, n_features_attn, 10)  # Using 10 components
pca_components_hidden = min(n_samples - 1, n_features_hidden, 50)  # Using 50 components

pca_attn = PCA(n_components=pca_components_attn)
attention_pca = pca_attn.fit_transform(attention_vectors)  # [20, 10]

pca_hidden = PCA(n_components=pca_components_hidden)
hidden_pca = pca_hidden.fit_transform(hidden_state_vectors)  # [20, 50]

print(f"   PCA explained variance (attention): {pca_attn.explained_variance_ratio_.sum():.2%}")
print(f"   PCA explained variance (hidden): {pca_hidden.explained_variance_ratio_.sum():.2%}")

# t-SNE (t-Distributed Stochastic Neighbor Embedding)
# It's for Non-linear reduction to 2D for visualization
# Preserves local structure (similar emails will stay close together)

perplexity = min(5, n_samples // 3)  # It controls local vs global structure balance

print("   Running t-SNE on attention patterns")
tsne_attention = TSNE(
    n_components=2,          # Reduce to 2D
    random_state=42,         
    perplexity=perplexity,   # Neighborhood size (lower value: focus on local structure)
    n_iter=1000,             # Optimization iterations
    learning_rate='auto',    # Adaptive learning rate
    init='pca'               # Initialize with PCA for stability
)
attention_2d = tsne_attention.fit_transform(attention_pca)  # [20, 2]

print("   Running t-SNE on hidden states")
tsne_hidden = TSNE(n_components=2, random_state=42, perplexity=perplexity,
                   n_iter=1000, learning_rate='auto', init='pca')
hidden_2d = tsne_hidden.fit_transform(hidden_pca)  # [20, 2]


print(f"   Using {pca_components_attn} PCA components for attention")
print(f"   Using {pca_components_hidden} PCA components for hidden states")
print("Dimensionality reduction is complete")
```
